```python
df.rename(columns={'Title': 'Game Title','Genre': 'Genre(s)', 'User Rating': 'User Rating (out of 100)'}, inplace=True) # renaming columns
print(df)
```

```python
# setting the index
df.set_index('Popularity', inplace=True)
df.index.name = 'index'

print(df.index)
```

```python
# import 2nd dataframe and set its index
df2 = pd.read_csv("imdb_games_cast.csv")
df2.set_index('Popularity', inplace=True)
df2.index.name = 'index'

# merge both dataframes (by index)
df_merged = df.merge(df2, left_index=True, right_index=True, how='inner')
pd.set_option('display.max_columns', 11) # update head & tail to display 11 columns
df_merged.head()

```

```python
# multiply 10 to each element in User Rating
df_merged['User Rating (out of 100)'] = df_merged['User Rating (out of 100)'].apply(lambda x: x * 10)
pd.set_option('display.float_format', '{:.0f}'.format) # do not display decimals
df_merged.head()
```

```python
# query df to filter rows where user rating > 95
rating_query = df_merged.query(f'`User Rating (out of 100)` > 95')
print(rating_query)
```

```python
# count NaN values
df_merged.isnull().sum()
```

```python
# group by genres and calculate their mean ratings using an aggregate function (mean)
ratings_by_genre = df_merged.groupby('Genre(s)')['User Rating (out of 100)'].mean()
print(ratings_by_genre)
```

```python
df_dropped = df_merged.drop_duplicates(subset=['Game Title'])
df_pivot = df_dropped.pivot(index='Game Title', columns='Genre(s)', values='User Rating (out of 100)')
df_pivot.head()
```

```python
# cleaning process
df_merged.drop('Runtime', axis=1, inplace=True) # drop runtime column since it is irrelevant and contains too many NaNs
df_merged.drop('Number of Votes', axis=1, inplace=True)

df_merged['Director'].fillna('unspecified', inplace=True) # substitute NaN values in director column with unspecified
df_merged['User Rating (out of 100)'].fillna('N/A', inplace=True)
df_merged['Year'].fillna('unspecified', inplace=True)
df_merged['Summary'].fillna('No information filled', inplace=True)
df_merged['Starring'].fillna('unspecified', inplace=True)
df_merged['Certificate'].fillna('unspecified', inplace=True)

df_merged.tail()



```

```python
import locale
import ast
import re
import json

df3 = pd.read_csv('movies_metadata.csv')

df3.head()
```

```python
df3.info()
```

```python
# CLEANING 'budget' and 'revenue'
# convert budget & revenue to numeric & add appropriate formatting
df3['budget'] = df3['budget'].replace('0', 'N/A')
df3['revenue'] = df3['revenue'].replace(0, 'N/A')
def format_with_commas(value):
    try:
        # check if the value can be converted to a float
        float_value = float(value)
        
        # format the float value with commas
        formatted_value = locale.format_string("$%.0f", float_value, grouping=True)
        
        return formatted_value
    
    except ValueError:
        # if the value is not a valid number, return it as is
        return value

# set the locale for formatting
locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')

# apply the custom function to the DataFrame columns
df3['budget'] = df3['budget'].apply(format_with_commas)
df3['revenue'] = df3['revenue'].apply(format_with_commas)

# reset the locale
locale.setlocale(locale.LC_ALL, 'C')



# CLEANING 'belongs_to_collection'
# convert to dict
df3['belongs_to_collection'] = df3['belongs_to_collection'].str.extract(r"'name': '([^']+)'")

# Rename the column to 'collection_name'
df3 = df3.rename(columns={'belongs_to_collection': 'collection_name'})



# CLEANING 'genres'
# convert to dict
df3['genres'] = df3['genres'].apply(ast.literal_eval)

# extract just the genre names and join them into a single string
df3['genres'] = df3['genres'].apply(lambda x: ', '.join(genre['name'] for genre in x))



# CLEANING 'production_companies', 'production_countries', and 'spoken_languages'
def extract_names(json_string):
    try:
        # Use ast.literal_eval to safely parse the JSON-like string
        parsed = ast.literal_eval(json_string)
        
        if isinstance(parsed, list):
            # Extract company names from the list of dictionaries
            names = [item['name'] for item in parsed]
            return ', '.join(names)
    except (SyntaxError, ValueError):
        pass  # Handle exceptions (e.g., for invalid JSON-like strings) by returning None
    
    # Handle NaN values and empty lists by returning 'unspecified'
    return 'unspecified'

# apply function to columns
df3['production_companies'] = df3['production_companies'].fillna('[]').apply(extract_names)
df3['production_countries'] = df3['production_countries'].fillna('[]').apply(extract_names)
df3['spoken_languages'] = df3['spoken_languages'].fillna('[]').apply(extract_names)



# CLEANING 'vote_average' and 'vote_count'
df3['vote_average'] = df3['vote_average'].apply(lambda x: float(x) if pd.notna(x) else None)
df3['vote_count'] = df3['vote_count'].apply(lambda x: f'{int(x):,}' if pd.notna(x) else None)

# drop unnecessary columns and fill NaNs
df3.drop(columns=['adult', 'video', 'original_language'], inplace=True)
df3['collection_name'].fillna('unspecified', inplace=True)
df3['homepage'].fillna('unspecified', inplace=True)
df3['id'].fillna('unspecified', inplace=True)
df3['imdb_id'].fillna('unspecified', inplace=True)
df3['overview'].fillna('unspecified', inplace=True)
df3['popularity'].fillna('N/A', inplace=True)
df3['poster_path'].fillna('unspecified', inplace=True)
df3['release_date'].fillna('unspecified', inplace=True)
df3['runtime'].fillna('unspecified', inplace=True)
df3['status'].fillna('unspecified', inplace=True)
df3['tagline'].fillna('N/A', inplace=True)
df3['title'].fillna('unspecified', inplace=True)
df3['vote_average'].fillna('N/A', inplace=True)
df3['vote_count'].fillna('N/A', inplace=True)
df3.info()
df3.to_csv('movies_metadata_updated.csv', index=False)



```

```python
# rearrange and rename columns
rearranged = [
    'id', 'imdb_id', 'title', 'original_title', 'collection_name', 'genres', 'production_companies', 'production_countries', 'release_date', 'spoken_languages', 'tagline', 'overview',
    'runtime', 'budget', 'revenue', 'vote_average', 'vote_count', 'popularity', 'homepage', 'poster_path', 'status'
]
df3 = df3[rearranged]

renamed = {
    'id': 'ID', 'imdb_id': 'IMDb ID', 'title': 'Title', 'original_title': 'Original Title', 'collection_name': 'Collection', 'genres': 'Genre(s)', 'production_companies': 'Production Company(s)',
    'production_countries': 'Production Country(s)', 'release_date': 'Release Date', 'spoken_languages': 'Languages', 'tagline': 'Tagline', 'overview': 'Overview', 'runtime': 'Runtime',
    'budget': 'Budget', 'revenue': 'Revenue', 'vote_average': 'Vote Average', 'vote_count': 'Vote Count', 'popularity': 'Popularity',  'homepage': 'Homepage', 'status': 'Status',
    'poster_path': 'Poster Path'
}
df3.rename(columns = renamed, inplace = True)
df3.head()
```
